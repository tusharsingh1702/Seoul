```python
import pandas as pd

# Load the datasets
df1 = pd.read_csv('1.csv')
df2 = pd.read_csv('2.csv')
df3 = pd.read_csv('3.csv')

df4 = pd.read_csv('4.csv')
df5 = pd.read_csv('5.csv')
df6 = pd.read_csv('6.csv')

df7 = pd.read_csv('7.csv')
df8 = pd.read_csv('8.csv')
df9 = pd.read_csv('9.csv')

df10 = pd.read_csv('10.csv')
df11 = pd.read_csv('11.csv')
df12 = pd.read_csv('12.csv')
# Concatenate the datasets
df= pd.concat([df1, df2, df3,df4,df5,df6,df7,df8,df9,df10,df11,df12])

# Save the merged dataset to a new CSV file
df.to_csv('merged_dataset.csv', index=False)
```

```python
# Convert the 'datetime' column to datetime format
df['datetime'] = pd.to_datetime(df['datetime'])

# Create 'Date' and 'Hour' columns
df['Date'] = df['datetime'].dt.date
df['Hour'] = df['datetime'].dt.hour

# Drop the original 'datetime' column if no longer needed
df.drop(columns=['datetime'], inplace=True)

# Reorder columns if necessary
df = df[['Date', 'Hour'] + [col for col in df.columns if col not in ['Date', 'Hour']]]

# Display the updated DataFrame
print(df)
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%Y-%m-%d')
df13['Date'] = pd.to_datetime(df13['Date'], format='%Y-%m-%d')

# Optional: Ensure 'Hour' columns are of the same type
df['Hour'] = df['Hour'].astype(int)  # Convert to integer if needed
df13['Hour'] = df13['Hour'].astype(int)

# Step 2: Merge the datasets on 'Date' and 'Hour' columns
merged_df = pd.merge(df, df13, on=['Date', 'Hour'], how='inner')  # Use 'inner' for only matching rows, 'outer' for all rows, etc.

# Display the first few rows of the merged dataset
print(merged_df.head())
```

```python
merged_df['Weekday'] = merged_df['Date'].dt.dayofweek < 5  # Monday to Friday are 0-4
merged_df['Weekday'] = merged_df['Weekday'].map({True: 'Yes', False: 'No'})
```
